```python
import numpy as mp
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
def initalize_centroids(x,k):
 indices=mp.random.choice(len(x),k,replace=True)
 return x[indices]
```

```python
def assign_to_clusters(x,centroids):
  dis=mp.linalg.norm(x[:,mp.newaxis] - centroids,axis=2)
  return mp.argmin(dis,axis=1)
```

```python
def update_centroids(x,labels,k):
  centroids=mp.array([x[labels == i].mean(axis=0) for i in range(k)])
  return centroids
```

```python
def k_means(x,k,max_iters=100):
  centroids=initalize_centroids(x,k)

  for _ in range(max_iters):
    labels=assign_to_clusters(x,centroids)
    new_centroids=update_centroids(x,labels,k)

    if mp.all(centroids==new_centroids):
      break

    centroids=new_centroids

  return labels, centroids
```

```python
mp.random.seed(42)
x=mp.concatenate([mp.random.normal(loc=i,scale=1,size=(50,2)) for i in range(5)])
x
```

```python
k=5
labels,centroids=k_means(x,k)
```

```python
plt.scatter(x[:,0], x[:,1], c=labels, cmap='viridis')
plt.scatter(centroids[:,0], centroids[:,1], marker='x',s=200, c='red',label='Centroids' )
plt.title('K-means CLustering')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```
